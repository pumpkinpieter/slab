```python
import numpy as np
import matplotlib as plt
from slab import SlabExact
from IPython.display import Video
```

```python
A = SlabExact(ns=[1.2, 1.5, 1.2], ts=[5e-6, 2e-6, 5e-6], xrefs=[400,200,400], symmetric=True)
```

```python
k_low=A.K_low
```

```python
field_type = 'TM'
mode_type = 'radiation'
beta = .99 * k_low
F = A.fields(beta, c0=1, field_type=field_type, mode_type=mode_type, sign='1')
A.plot_field_1d(F, plot_axis=True, part='real', plot_regions=True, contrast=.1, hatch=None)
```

```python
A.radiation_mode_phase_animation('radiation_phase_at_klow', k_low, secs=3, field_type=field_type, 
                                 plot_regions=True, contrast=.1, hatch=None)
```

```python
Video('radiation_phase_at_klow'+'.mp4', html_attributes="controls loop autoplay")
```
