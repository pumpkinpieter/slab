# Branch Cuts in Psi based fields

Are there any?  And do they persist to the spectral integrand?

```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact, plot_complex, plot_complex_surface
plt.style.use('dark_background')
%matplotlib widget
```

```python
A = SlabExact(ns=[1.4, 1.43, 1.4], Ts=[3, 3, 3], symmetric=True, wl=2e-6)
```

There are (it seems unavoidable) branch cuts when determining the propagation constants in one region ($R_i$) from those in another reference region ($R_0$) regardless of whether these constants are the $Z_i$ or the angles $\Psi_i$.  

```python
plot_complex(A.Zi_from_Z0, -np.pi/1.5, np.pi/1.5, -2, 2, fargs=(A.ns[1],), part='real', log_abs=False)
```

```python
plot_complex(A.Psi_i_from_Psi0, -np.pi/1.5, np.pi/1.5, -1, 1, fargs=(A.ns[1],), part='imag', log_abs=False)
```

These cuts persist in the fields:

```python
plot_complex(A.evaluate_fields, -5, 5, -2, 2, fargs=(0,), fkwargs={'Normalizer':A.normalizer('ours')}, part='imag', 
            log_abs=False, )
```

However, they typically disappear in the spectral integrand.

```python
plot_complex_surface(A.spectral_integrand, -1.3*A.K0, 1.1*A.K0, -2, 2, fkwargs={'s':0, 'z':2, 'Normalizer':A.normalizer('eigvec')},
                     iref=200, rref=200, part='real', max_value=2, min_value=-.5, z_lims=(-1, 4))
```
