# Sign of Zs when input Z0 is negative

What should the convention be?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton
from slab import SlabExact
plt.style.use('dark_background')
%matplotlib widget
```

```python
A = SlabExact(scale=1e-5, Ts=[6.5, 1, 8, 2.5, 6.5],
      ns=[1, 1.44, 1, 1.44, 1], xrefs=[100, 40, 300, 40, 100],
      symmetric=True, wl=6.9e-6)

B = SlabExact(scale=1e-5, Ts=[6.5, 1, 8, 2.5, 6.5],
      ns=[2, 1.44, 2, 1.44, 2], xrefs=[100, 40, 300, 40, 100],
      symmetric=True, wl=6.9e-6)
```

```python
Zs = np.array([[4+1j, 4-1j], [3, -1]])
```

```python
A.Zi_from_Z0(Zs, A.ns)
```

```python
def Zi_from_Z0(slab, Z0, ni):
    '''Get Z on region i (with refractive index ni) from Z0. Vectorized
    in Z0 and ni'''
    Z0, ni = np.array(Z0), np.array(ni)
    return (np.sign(Z0[..., np.newaxis]) * np.sqrt((ni[np.newaxis]**2 - slab.n0**2) * slab.K0**2 +
                   Z0[..., np.newaxis]**2, dtype=complex))
    return Zs
```
