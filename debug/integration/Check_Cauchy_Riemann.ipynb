# Checking Cauchy Riemann equations

To see if Cauchy's theorem should hold

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def check_CR(f, z, h=1e-8):
    '''Check if f is analytic at z using width h.'''
    fz = f(z)
    f_re_h = f(z + h)
    f_im_h = f(z + 1j * h)
    dudx, dudy = (f_re_h - fz).real / h, (f_im_h - fz).real / h
    dvdx, dvdy = (f_re_h - fz).imag / h, (f_im_h - fz).imag / h
    return dudx - dvdy, dudy + dvdx
```

```python
def f(z):
    return np.conj(z)
```

```python
check_CR(f, 1)
```
