# Basic Trapezoid rule method implementation

```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
```

```python
def f(x):
    return np.sin(x)
```

```python
a = 0
b = np.pi
N = 100001

xs = np.linspace(a, b, N)
dx = xs[1:] - xs[:-1]

integral = f(xs[0]) * dx[0]
integral += np.sum([f(xs[i]) * (dx[i] + dx[i-1]) for i in range(1, N-1)])
integral += f(xs[-1]) * dx[-1]
integral *= 1/2
```
