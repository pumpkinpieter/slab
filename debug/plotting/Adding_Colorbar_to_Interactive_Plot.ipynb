# Adding Colorbar to Interactive Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
from ipywidgets import interactive, FloatSlider, Layout

%matplotlib widget
plt.style.use('dark_background')
```

```python
plt.close('all')
fig, (ax, cbar_ax) = plt.subplots(1,2, figsize=(12, 6), width_ratios=[10,1])

rmin, rmax, rref = -1,1,100
imin, imax, iref = -1,1,100
xs, ys = np.linspace(rmin, rmax, rref), np.linspace(imin, imax, iref)

Xs, Ys = np.meshgrid(xs, ys)
Zs = Xs + 1j * Ys

minwl=1
maxwl=2
startwl = 1.5

Nwl = 100
levels=50

def F(wl):
    return wl * (Xs**2 + np.sin(Ys))

fs = F(startwl)
data = ax.contour(Xs, Ys, fs, levels=levels)
cbar = plt.colorbar(data, ax=cbar_ax)

def det_plot(wl):
    ax.clear()
    fs = F(wl)

    data = ax.contour(Xs, Ys, fs, levels=levels)

    ax.grid(True)
    ax.set_facecolor('grey')
    ax.set_title('Wavelength: %.6e\n' % (wl), fontsize=14)
    cbar = plt.colorbar(data, ax=cbar_ax)
    cbar.remove()    

    plt.show()

step = (maxwl - minwl) / Nwl

interactive_plot = interactive(det_plot,
                               wl=FloatSlider(min=minwl, max=maxwl,
                                              step=step, value=minwl,
                                              readout_format='.5e',
                                              layout=Layout(
                                                  width='50%')))
output = interactive_plot.children[-1]
output.layout.width = '2000px'
interactive_plot
```
