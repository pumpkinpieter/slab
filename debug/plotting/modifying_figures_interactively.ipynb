# Add an axis

## Via add_axes

```python
fig.add_axes?
```

```python
plt.close('all')
fig, ax = plt.subplots(1, figsize=(5,5));
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
y2 = np.cos(x)
ax.plot(x,y);


# top=0.88,
# bottom=0.11,
# left=0.125,
# right=0.9,
# hspace=0.2,
# wspace=0.2

ax2 = fig.add_axes((1, .11, .5, 1))
ax2.plot(x,y2);
```

```python
# Create figure() objects 
# This acts as a container  
# for the different plots 
plt.close('all')
fig = plt.figure(figsize=(4,4))  
  
# Creating two axes 
# add_axes([xmin,ymin,dx,dy]) 
ax1 = fig.add_axes([1.2,1,1.2,1])
# ax = plt.gca()
ax1.plot(x,y)
ax2=fig.add_axes([0,1,1,1]) 
ax2.plot(x,y2)
```

```python
# Create figure() objects 
# This acts as a container  
# for the different plots 
plt.close('all')

fig = plt.figure(figsize=(3,2))  
  
# Creating two axes 
# add_axes([xmin,ymin,dx,dy]) 
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
# ax1=fig.add_axes([0,1,1,1]) 
# ax1.plot(x,y)
```

## Via subplots

```python
fig.add_subplot?
```

```python
plt.close('all')

fig, ax1 = plt.subplots(1)
# ax = plt.gca()
# ax1 = fig.add_subplot(122)
ax2 = fig.add_subplot(121, sharey=ax1)
# ax = fig.add_subplot(246)

ax1.plot(x,y)
ax2.plot(x,y2)
fig.subplots_adjust(wspace=.31)
# plt.subplot_tool()
# top=0.88,
# bottom=0.11,
# left=0.125,
# right=0.9,
# hspace=0.2,
# wspace=0.5
```

```python
plt.close('all')

fig, ax1 = plt.subplots(1)
ax1.plot(x,y)

fig.add_gridspec(ncols=2, nrows=1)
ax2 = fig.add_subplot(121)
```
