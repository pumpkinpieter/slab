```python
%matplotlib widget
from ipywidgets import interactive, Play, IntSlider, jslink, HBox, FloatSlider
import matplotlib.pyplot as plt
import numpy as np
plt.figure(1)

def f(m, b):
    ax = plt.gca()
    ax.clear()
    x = np.linspace(-10, 10, num=1000)
    ax.plot(x, m * x + b)
    ax.set_ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2, 2, .01), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '650px'
interactive_plot
```

```python
plt.figure(2)

def f(m, b):
    ax = plt.gca()
    ax.clear()
    x = np.linspace(-10, 10, num=1000)
    s = m * 1/50
    ax.plot(x, s * x + b)
    ax.set_ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=Play(min=-200, max=200, interval=100), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '650px'
interactive_plot
```
