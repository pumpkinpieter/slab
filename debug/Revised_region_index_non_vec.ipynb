```python
import numpy as np
from slab import SlabExact
```

```python
A = SlabExact()
```

```python
def region_index(x, continuity='left'):
    for j, Rho in enumerate(A.Rhos[1:]):
        if continuity == 'left':
            if x <= Rho:
                return j
        elif continuity == 'right':
            if x < Rho:
                return j
        else:
            raise ValueError('Continuity must be set to left or right.')
    return j
```

```python
region_index(3.5, continuity='left')
```
