```python
import numpy as np
import json 

a = np.arange(10).reshape(2,5) # a 2 by 5 array
b = a.flatten().tolist() # nested lists with same data, indices
with open('test.json', 'w') as outfile:
    
    json.dump(b, outfile, 
              separators=(',', ':'), 
              sort_keys=True, 
              indent=4) ### this saves the array in .json format
```
