```python
import numpy as np
from slab import SlabExact
from fiberamp.fiber.microstruct.bragg import plotlogf, plotlogf_real
from scipy.optimize import newton
```

```python
A = SlabExact(symmetric=True)
```

```python
field_type = 'TM'
mode_type = 'guided'
```

```python
plotlogf(A.determinant, 0, A.K_high, -3, 3, field_type, mode_type, None, 'standard', 'imag', 
         iref=100, rref=100,levels=100)
```

```python
plotlogf_real(A.determinant,2.2,2.3, field_type, mode_type, None, 'standard', 'imag')
```

```python
beta1 = newton(A.determinant, 2.23, args=('TM', mode_type, None, 'standard', 'imag'))
```

```python
A.coefficients(beta1).T
```

```python
F = A.fields(beta1, field_type=field_type, mode_type=mode_type)
```
