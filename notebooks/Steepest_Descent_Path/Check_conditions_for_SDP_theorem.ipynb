```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
%matplotlib widget
plt.style.use('dark_background')
```

```python
A = SlabExact(symmetric=True, wl=1e-5)
```

```python
Zs = A.sdp_contour(-5, 5, 1000)['Zs']
```

```python
Betas = A.Beta_from_Z(Zs)
Re_i_Betas = - Betas.imag
Im_i_Betas = Betas.real
```

```python
plt.close('all')
plt.plot(Zs.real, Zs.imag)
plt.plot(Zs.real, Re_i_Betas)
plt.plot(Zs.real, Im_i_Betas)
```
