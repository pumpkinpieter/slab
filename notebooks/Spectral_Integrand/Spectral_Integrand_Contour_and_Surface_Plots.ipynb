# Functions for Visualizing Spectral Integrand

These plot the spectral integrand as contour plots and surfaces (in Z plane).  If no input field ($f_0$) is provided, it assumes a delta source.

```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
plt.style.use('dark_background')
%matplotlib widget
```

```python
A = SlabExact(scale=1e-5, Ts=[6.5, 1, 1, 1, 6.5],
              ns=[1, 1.44, 1, 1.44, 1], xrefs=[100, 40, 300, 40, 100],
              symmetric=True, no_xs=True, wl=2e-5)
```

```python
Normalizer = A.normalizer('eigvec')
```

```python
A.spectral_integrand_plot(-.1, 4, -2, 2,
                          rref=200, iref=120,
                          s=1, z=0, x=-1.5, 
                          log_abs=False, part='real',
                          Normalizer=Normalizer,
                          facecolor='k', grid=False,
                          max_z=1, min_z=-1,
                         )
```

```python
A.spectral_integrand_surface_plot(-.1, 4, -2, 2,
                                  rref=200, iref=120,
                                  s=1.4, z=0, x=-1.5, 
                                  max_z=1, min_z=-1,
                                  log_abs=False, part='real',
                                  elev=20, zoom=3,
                                  Normalizer=Normalizer);
```
