# Circular contour integration in complex plane

Preparatory notebook for normalizing leaky modes via contour integration.

```python
import numpy as np
```

```python
def contour_int(f, c, r, N=10):
    ts = np.linspace(0, 2*np.pi, N+1)[:-1]
    dt = 2*np.pi / N
    dgamma = 1j * r * np.exp(1j * ts)
    fs = f(c + r * np.exp(1j * ts))
    return np.sum(fs * dgamma * dt)
```

```python
f = lambda z: np.cos(z) / z
```

```python
def evalZ(F, Z):
    Fs = np.zeros(len(Z), dtype=object)
    for i in range(len(Z)):
        Fs[i] = lambda x, z: F(x, z, Z[i])
    return Fs
```
