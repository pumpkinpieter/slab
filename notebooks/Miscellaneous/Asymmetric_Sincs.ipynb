```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
nl = 1
nr = 2
k0 = 1

vsqr = k0**2*(nr**2 - nl**2)

def sinc_asym(Z1r, R=1, Z2r=1.1*vsqr):
    Z1l = np.sqrt(Z1r**2 - vsqr, dtype=complex)
    Z2l = np.sqrt(Z2r**2 - vsqr, dtype=complex)
    return np.sin(R*(Z1l - Z2l))/(Z1l - Z2l)
    
```

```python
xs = np.linspace(vsqr,10, 10000)
ys = sinc_asym(xs, R=50)
```

```python
plt.close('all')
plt.plot(xs, ys.real)
plt.plot(xs, ys.imag)
```

```python
xs = np.linspace(0, vsqr, 1000)
ys = sinc_asym(xs, R=50)
```

```python
plt.close('all')
plt.plot(xs, ys.real)
plt.plot(xs, ys.imag)
```

Right, it looks like it's unbounded as R increases for area below vsqr.  So maybe we don't want to integrate using the Z associated to the higher of the two n values, maybe we want the smaller.  Or maybe we want to separate the integral, and then investigate the orthogonality of one sided radiation modes with guided and two sided radiation modes.
