```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
%matplotlib widget
# plt.style.use('dark_background')
from IPython.display import Video
```

```python
# A = SlabExact(ns=[2, 1], ts=[6e-6, 6e-6], xrefs=[1000, 1000], wl=2*np.pi*1e-6, symmetric=True)

A = SlabExact(ns=[1.7, 2, 1], ts=[5e-6, 2e-6, 5e-6], xrefs=[1000, 1000, 1000], wl=2e-6, symmetric=True)

# A = SlabExact(ns=[2, 1.8, 1.5, 1.3], ts = [1e-6, 2e-6, 2e-6, 1e-6], xrefs=[100, 100, 100, 100], scale=1e-6, symmetric=True)  
```

```python
A.plot_refractive_index(plot_regions=True, hatch=None)
```

```python
A.Z_one_sided,
```

```python
mode_type='radiation'
field_type='TE'
```

```python
Fe = A.fields(4, mode_type=mode_type, field_type=field_type, Z_plane=True, sign='-1')  # Odd mode just use phase=pi/2
Fo = A.fields(5, mode_type=mode_type, field_type=field_type, Z_plane=True, sign='-1')  # Odd mode just use phase=pi/2
```

```python
A.K_low, A.K_high
```

```python
M = A.transmission_matrix(2.5j,  Ztype_far_left='standard', Ztype_far_right='standard')
```
