# Studies of even modes

Slab even modes for symmetric structures don't eliminate cos - term directly in coefficient system.  Here we look at the function that gets integrated against cos -

```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
from fiberamp.fiber.microstruct.bragg import plotlogf, plotlogf_real
```

```python
A = SlabExact(ns=[1, 2, 1], symmetric=True)  # Need symmetric interfaces for properties to hold.
```

```python
betas = np.linspace(0.0001, 100, 2000)
ys = np.zeros_like(betas, dtype=complex)
```

```python
Cs = A.coefficients(betas, mode_type='radiation', sign='1', Z_plane=True)
```

```python
S = np.array([[0,1],[1,0]])
J = np.array([[0,2j],[-2j, 0]])
```

```python
I = 100
v0, w0 = Cs[I].T[0], Cs[I].T[-1]
```

```python
for i in range(len(Cs)):
    v, w = Cs[i].T[0], Cs[i].T[-1]
    ys[i] = v.T @ J @ v0 - w.T @ J @ w0
```

```python
plt.plot(betas, ys)
```
