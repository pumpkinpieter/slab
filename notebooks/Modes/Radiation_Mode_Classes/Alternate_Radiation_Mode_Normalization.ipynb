```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
from scipy.linalg import eig
from fiberamp.fiber.microstruct.bragg import plotlogf, plotlogf_real
from scipy.optimize import newton
from IPython.display import Video
```

```python
A = SlabExact(ns=[1.4, 1.5, 1.4], ts=[1e-6, 1e-6, 1e-6], wl=.9e-6,
              symmetric=True) # Need symmetric interfaces for properties to hold.
```

```python
A.plot_refractive_index()
```

```python
F = A.fields(10, Z_plane=True, paper_method=False, mode_type='radiation', sign='-1')
```

```python
A.plot_field_1d(F, xs=np.linspace(-5,5,1000), part='real')
ax = plt.gca()
# ax.set_ylim(-1,1)
```

```python
A.radiation_mode_Z_animation('Z_anim', Z_range=[-A.Z_evanescent, A.Z_evanescent], paper_method=False, sign='-1')
```

```python
Video('Z_anim'+'.mp4', html_attributes="controls loop autoplay")
```
