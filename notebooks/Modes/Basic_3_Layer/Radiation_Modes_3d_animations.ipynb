# Radiation mode animations

Again for the simple 3 layer structure: cladding-core-cladding.  

```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
from fiberamp.fiber.microstruct.bragg import plotlogf, plotlogf_real
from scipy.optimize import newton
from IPython.display import Video
# plt.style.use('dark_background')
%matplotlib widget
```

```python
A = SlabExact(ns=[1.1, 1.5, 1.1], Ts=[5e-6, 2e-6, 5e-6], xrefs=[125, 50, 125], symmetric=True)
```

```python
field_type = 'TE'
mode_type = 'radiation'
```

```python
beta = 0.9*A.K_low
Z_prop  = A.Z_from_Beta(beta)
Z_ev = 1.01*A.Z_evanescent
```

```python
F_standing = A.fields_Z(Z_prop, field_type=field_type, mode_type=mode_type, sign='1', c0=1, c1=1)
F_traveling = A.fields_Z(Z_prop, field_type=field_type, mode_type=mode_type, sign='1', c0=.00, c1=1)
```
