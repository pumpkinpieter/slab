# Guided modes of Hole Waveguide?

Checking because we found a sneaky guided mode in the hollow core case, and propagation hasn't fully gotten us there with hole waveguide



```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact
from fiberamp.fiber.microstruct.bragg import plotlogf, plotlogf_real
from scipy.optimize import newton
plt.style.use('dark_background')
%matplotlib widget
```

```python
A = SlabExact(scale=1e-6, ts=[4e-6, 2e-6,4e-6],
              ns=[1.44, 1, 1.44], xrefs=[400, 200, 400],
              symmetric=True, wl=.55e-6)
```

```python
# Set input field type
field_type = 'TE'
```

```python
mode_type = 'guided'
k_low = A.K_low
```

```python
plotlogf(A.determinant, .8*k_low, 1.1*k_low, -.2, .2, field_type, mode_type, iref=100, rref=100,levels=100,
         figsize=(10,5))
```

```python
beta1 = newton(A.determinant, np.array([11.4]), args=('TE', mode_type))
print(abs(A.determinant(beta1, 'TE', mode_type)))
beta1
```
