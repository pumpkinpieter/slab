#  Interactive plot of leaky mode locations as function of wavelength

We can see the behavior as we approach the resonance frequency and move past it.

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
from slab import SlabExact
%matplotlib widget
```

```python
loss_min_wl = 5.828070175438596e-06
max_loss_wl = 6.90764472e-06
```

```python
A = SlabExact(scale=1e-5, Ts=[6.5, 1, 8, 1, 6.5],
              ns=[1, 1.44, 1, 1.44, 1], xrefs=[100, 40, 300, 40, 100],
              symmetric=True, wl=max_loss_wl)
```

```python
A.interactive_determinant_plot(minwl=.98*max_loss_wl, maxwl=1.02*max_loss_wl, rmin=-.4, rmax=4, imin=-.5, imax=.4, 
                               sdp_sign=-1, levels=70, iref=200, rref=200, Nwl=1000,)
```
