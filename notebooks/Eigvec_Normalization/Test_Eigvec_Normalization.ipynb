# Test eigenvector normalization now that we debugged

```python
import numpy as np
import matplotlib.pyplot as plt
from slab import SlabExact, plot_complex
from scipy.optimize import newton
plt.style.use('dark_background')
%matplotlib widget
```

```python
A = SlabExact(scale=1e-5, Ts=[6.5, 1, 8, 2.5, 6.5],
              ns=[1, 1.44, 1, 1.44, 1], xrefs=[100, 40, 300, 200, 100],
              symmetric=True, wl=6.9e-6)
```

```python
Eig = A.normalizer('eigvec')
Ours = A.normalizer('ours')
Paper = A.normalizer('paper')
```

```python
Fr = A.fields(1.3, Normalizer=Eig, mode_type='radiation', sign=-1)
A.plot_field_1d(Fr)
```

```python
A.spectral_integrand(1, s=1, x=1)
```

```python
A.spectral_integrand_plot(-.8, .8, -.3, .3, s=1, Normalizer=Ours)
```
