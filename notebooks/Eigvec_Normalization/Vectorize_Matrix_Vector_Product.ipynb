# Vectorize application of matrix to vector

So that we can apply the transfer matrix iteratively to an array of vectors.  I have done this before, but kind of didn't like how I did it.  However, it looks like I did it right, there's no better way I find below.

```python
import numpy as np
```

```python
Cs = np.array([[1,2,3], [1,2,3]])

M = np.zeros(Cs.shape + (2,2))
M[..., 0,0] = Cs
M[..., 0,1] = np.sin(Cs)
M[..., 1,0] = Cs / 2
```

```python
M[0]
```

```python
V = np.zeros(Cs.shape + (2, 1))
V[..., 0, 0] = Cs
V[..., 1, 0] = -Cs
```

```python
M[0] @ np.ones((2, 1))
```

```python
(M @ V)[0,0]
```

```python
M[0,0] @ V[0,0]
```

One thing that might be useful though is doing the row vector version

```python
Vr = np.zeros(Cs.shape + (1, 2))
Vr[..., 0, 0] = 2*Cs
Vr[..., 0, 1] = -.5*Cs
Vr
```

```python
Vr @ M
```
