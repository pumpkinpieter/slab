```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad_vec
%matplotlib widget
```

```python
xs = np.linspace(-np.pi, np.pi, 1000)
Omegas = np.array([[.25, 2.5, 25]])
ys = np.cos(xs) * np.exp(-Omegas.T * xs**2)
ys.shape
```

```python
plt.close('all')
fig, ax = plt.subplots(1, figsize=(10,4))
ax.axhline(0, lw=.5, c='k')
# ax.axvline(0, lw=.5, c='k')
# ax.grid('minor')
for i, data in enumerate(ys):
    ax.plot(xs, data, color=str((i/len(ys))), lw=1.2, label='$\Omega = %.2f$'% Omegas.T[i])
plt.legend()
ax.set_xlabel('z')
fig.savefig('cos_example_integrand.jpg', dpi=600)
```

```python
def F(x, Omegas=np.linspace(0, 100, 1000)):
    return np.cos(x) * np.exp(-Omegas.T * x**2)
```

```python
omegas = np.linspace(0, 15, 1000)
ys2 = quad_vec(F, -np.pi, np.pi, args=(omegas,))[0]
```

```python
plt.close('all')
fig, ax = plt.subplots(1, figsize=(10,4))
ax.plot(omegas, ys2, label='$I(\Omega)$', color='k')
xs = np.linspace(.01, 15, 300)
ax.plot(xs, np.sqrt(np.pi / xs), color='gray', lw=1, ls='--',
        label='$\\sqrt{\\frac{\pi}{\Omega}}$')
plt.legend(fontsize=12)
ax.axhline(0, lw=.8, c='k')
ax.axvline(0, lw=.8, c='k')
ax.set_xlabel('$\Omega$', fontsize=12)
ax.set_ylim(-.2, 2.2)
fig.savefig('cos_example_asymptotics.jpg', dpi=600)
```

```python
def G(x, Omegas=np.linspace(0, 100, 1000)):
    return np.sin(x) * np.exp(-Omegas * x**2)
```

```python
omegas = np.linspace(0, 15, 1000)
ys3 = quad_vec(G, 0, np.pi, args=(omegas,))[0]
```

```python
plt.close('all')
fig, ax = plt.subplots(1, figsize=(10,4))
ax.plot(omegas, ys3, label='$I(\Omega)$', color='k', lw=.8)
# xs = np.linspace(.01, 15, 300)
# ax.plot(xs, np.sqrt(np.pi / xs), color='gray', lw=1, ls='--',
#         label='$\\sqrt{\\frac{\pi}{\Omega}}$')
plt.legend(fontsize=12)
ax.axhline(0, lw=.8, c='k')
ax.axvline(0, lw=.8, c='k')
ax.set_xlabel('$\Omega$', fontsize=12)
ax.set_ylim(-.2, 2.2)
fig.savefig('sin_example_asymptotics.jpg', dpi=600)
```
